## Let's try to answer some questions from this data

## Q1.Which city do people often travel to from Bangalore??

```python
# Filter DataFrame to not include Bengaluru as arrival city
df1 = df[df['Arrival_City'] != 'Bengaluru']
most_travelled = df1['Arrival_City'].value_counts()
most_travelled
```

```python
plt.figure(figsize=(15,8))
most_travelled.plot(kind = 'bar')
plt.show()
```

#### From above graphs we see that highest number of flights fly to Mumbai from Bangalore.

## Q2.Which Airline is usually late (Arrival or Departure)?

```python
# Filter DataFrame to include only flights that are delayed
delayed_flights = df[df['Flight_Time_status'] == 'Delayed']

# Group by 'Airlines Name' and count occurrences of delayed flights for each airline
delayed_flights_count = delayed_flights.groupby('Airlines_Name').size().reset_index(name='delayed_flights_count').sort_values(by='delayed_flights_count', ascending=False).head(5)

print("The top 5 Airlines that are delayed are:")
print(delayed_flights_count)
```

```python
plt.figure(figsize=(15, 8))
sns.stripplot(x="Flight_delay_by_time", y="Airlines_Name", data=df, jitter=0.05)
plt.title("Airlines vs Delayed time")

# Rotate x-axis labels perpendicular to the axis
plt.xticks(rotation=90)
plt.show()
```

### IndiGo has highest no. of delayed flights

## Q3.Which flight (FlightNo) is usually arriving late?

```python
# Filter DataFrame to include only arrival flights that are delayed
delayed_arrival_flights = df[(df['Flight_Status'] == 'Arrived') & (df['Flight_Time_status'] == 'Delayed')]

# Group by 'Flight_No' and count occurrences of delayed arrival flights for each aircraft
delayed_arrival_flights_count = delayed_arrival_flights.groupby('Flight_No').size().reset_index(name='Delayed_Arrival_Count').sort_values(by='Delayed_Arrival_Count', ascending=False).head(5)

print("The top 5 Aircrafts that are delayed on arrival are:")
print(delayed_arrival_flights_count)
```

```python
plt.figure(figsize=(8, 8))
delayed_arrival_flights_count.plot(kind='bar', x='Flight_No', y='Delayed_Arrival_Count', 
                                          legend=None, color='skyblue')
plt.title('Number of Delayed Arrival Flights by Aircraft')
plt.xlabel('Aircraft')
plt.ylabel('Number of Delayed Arrival Flights')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```

#### From above plot we observe that FlightNo. 6E 6354 has arrived late the highest no of times.

## Q6.Which flight took most time to cover and from where?

```python
Max_time = df['Flight_time'].max()

Max_time_flight = df.loc[df['Flight_time'] == Max_time, 'Arrival_City'].iloc[0]

print("Flight from ", Max_time_flight, " took maximum time of ", 
      Max_time, " to reach Banglore.")
```
